#### Bivariate Plot: <br>
*Definition*: A bivariate plot involves `two variables`. It shows the relationship between them. <br>
*Example*: Scatter plot, line plot, etc.

#### Multivariate Analysis:
*Definition*: A multivariate plot involves `three or more variables` to analyze their relationships simultaneously.<br>*Example*: a multivariate Analysis can show the relationship between two numerical variables and a categorical variable.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
titanic = pd.read_csv('train.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

## 1. Scatterplot (Numerical - Numerical)

#### Bivariate 

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

#### Multivate : <br>
>Scatter plot with hue, style, and size

```python
# Scatter plot with hue, style, and size
sns.scatterplot(x='total_bill', y='tip', hue='smoker', style='sex', size='size', data=tips)
plt.show()
```

## 2. Bar Plot (Numerical - Categorical)

```python
titanic.head()
```

#### Bivariate :

```python
sns.barplot(x='Pclass', y='Age', data=titanic)
plt.show()
```

#### Multivariate :

```python
sns.barplot(x='Pclass', y='Age', hue='Sex', data=titanic)
plt.show()
```

## 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(x='Sex', y='Age', data=titanic)
plt.show()
```

```python
sns.boxplot(x='Age', y='Sex', hue='Survived', data=titanic)
plt.show()
```

## 4. Distplot (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'], hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'], hist=False)
plt.show()
```

## kdeplot (Numerical - Categorical)

```python
# Plot the Age distribution for those who didn't survive (Survived == 0)
# Plot the Age distribution for those who survived (Survived == 1)
sns.kdeplot(titanic[titanic['Survived'] == 0]['Age'], label='Did not survive', fill=True)
sns.kdeplot(titanic[titanic['Survived'] == 1]['Age'], label='Survived', fill=True)

plt.show()
```

## 5. HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Sex').mean()['Survived']
```

## 6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

## 7. Pairplot

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

## 8. Lineplot (Numerical - Numerical)

```python
flights.head()
```

```python
new = flights.groupby('year')['passengers'].sum().reset_index()
```

```python
sns.lineplot(x='year',y='passengers',data=new)
plt.show()
```
